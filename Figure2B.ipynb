```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df2 = pd.read_csv(r'/media/shuang/data/Covariance/results/hist.csv')
```

```python
# plot hist for gradients of two groups
plt.figure(figsize=(4,6))
ax=sns.histplot(data=df2, x="G1", hue="group2",palette={2:'#4d4d4d', 1:'#33a02c'},alpha=0.5,bins=20,edgecolor='w', linewidth=1)
ax.set(xlim=(-0.04, 0.14))
plt.savefig("/media/shuang/data/Covariance/Figures/G1Hist.svg", format="svg",dpi=1000)
```

```python
plt.figure(figsize=(4,6))
ax=sns.histplot(data=df2, x="G2", hue="group2",palette={2:'#4d4d4d', 1:'#33a02c'},alpha=0.5,bins=20,edgecolor='w', linewidth=1)
ax.set(xlim=(-0.06, 0.07))
plt.savefig("/media/shuang/data/Covariance/Figures/G2Hist.svg", format="svg",dpi=1000)
```

```python
df = pd.read_csv(r'/media/shuang/data/Covariance_RAWg/results/N7.csv')
```
