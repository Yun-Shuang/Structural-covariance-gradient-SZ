```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'/media/shuang/data/Covariance/results/N7.csv')
```

```python
sns.regplot(data=df,x="conn_HC", y="GD",scatter=False,color='r', fit_reg=True)
sns.kdeplot(data=df,x="conn_HC", y="GD", cmap="Greys", shade=True,alpha=1)
plt.savefig("/media/shuang/data/Covariance/Figures/connGDcorr.svg", format="svg",dpi=1000)
```

```python
sns.regplot(data=df,x="HCG2", y="meanGD",scatter=False,color='r', fit_reg=True)
sns.kdeplot(data=df,x="HCG2", y="meanGD", cmap="Greys", shade=True,alpha=1)
plt.savefig("/media/shuang/data/Covariance/Figures/G2GDcorr.svg", format="svg",dpi=1000)
```

```python
sns.regplot(data=df,x="HCG1", y="meanGD",scatter=False,color='r', fit_reg=True)
sns.kdeplot(data=df,x="HCG1", y="meanGD", cmap="Greys", shade=True,alpha=1)
plt.savefig("/media/shuang/data/Covariance/Figures/G1GDcorr.svg", format="svg",dpi=1000)
```

```python
df2 = pd.read_csv(r'/media/shuang/data/Covariance_RAWg/results/BinCD.csv')
```

```python
plt.figure(figsize=(2,3))
sns.set_theme(style="whitegrid", palette="muted")
# Draw a categorical scatterplot to show each observation
ax = sns.swarmplot(data=df2, x="Group", y="TDEOSCDbin1")
sns.boxplot(x="Group", y="TDEOSCDbin1",data=df2, 
    notch=True, showcaps=False,
    flierprops={"marker": "x"},
    boxprops={"facecolor": (.3, .5, .7, .5)},
    medianprops={"color": "r", "linewidth": 2},
)
plt.savefig("/media/shuang/data/Covariance/Figures/GDBIN1.svg", format="svg",dpi=1000)
```
